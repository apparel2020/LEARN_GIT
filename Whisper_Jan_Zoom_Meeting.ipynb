```python
import whisper

model = whisper.load_model("large")  # Best for long audio & non-English languages
result = model.transcribe("Jan Zoom Meeting_24_03_2025.mp3", language="de")  # "de" = German
print(result["text"])
```

```python
import whisper

model = whisper.load_model("large")
result = model.transcribe("Jan Zoom Meeting_24_03_2025.mp3", language="de", verbose=True, task="transcribe")

# Save subtitles in SRT format
with open("transcription.srt", "w", encoding="utf-8") as f:
    f.write(whisper.utils.get_writer("srt", ".")("transcription.srt", result))

# Download the SRT file
from google.colab import files
files.download("transcription.srt")
```

```python
import whisper

model = whisper.load_model("large")
result = model.transcribe("Jan Zoom Meeting_24_03_2025.mp3", language="de", verbose=True, task="transcribe")

# Save subtitles in SRT format
with open("transcription.srt", "w", encoding="utf-8") as f:
    # The get_writer function should only take the filename as argument
    # and it returns a writer function
    writer = whisper.utils.get_writer("srt", ".")
    # The writer function should then be called with the filename and result
    f.write(writer(result=result)) # Pass result as a keyword argument

# Download the SRT file
from google.colab import files
files.download("transcription.srt")
```
